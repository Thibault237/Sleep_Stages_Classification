## Data

The goal is to classify a 30 seconds sample of biometric datas (7 electroencephalograms, one tri-dimensional accelerometer and one pulse sensor) into one of the following 4 sleep stages: Wake, N1, N2, N3, REM

For this purpose, we have a Dataset of 38289 samples annotated by experts.
Below is a training example labelled as stage 2 sleep

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import pandas as pd
%matplotlib inline

train_y = pd.read_csv('Data/train_y.csv',index_col=0)
```

```python
sampling_freq = {'accelerometer_x':10,'accelerometer_y':10,'accelerometer_z':10,
                      'eeg_1':50,'eeg_2':50,'eeg_3':50,'eeg_4':50,'eeg_5':50,'eeg_6':50,'eeg_7':50,
                      'pulse_oximeter_infrared':10}

def get_sample(sample):
    channels={}
    with h5py.File('Data/train.h5') as f:
        for key in sampling_freq:
            channels[key] = f[key][sample,:]
            
    return channels
```

```python
sampling1 = np.linspace(0,30,300)
sampling2 = np.linspace(0,30,1500)

samplings={10:sampling1,50:sampling2}

sample_num = 50
channels = get_sample(sample_num)

print('Sample labelled as {}'.format(train_y.iloc[sample_num]))
f, axes = plt.subplots(11,1,figsize=(13,23))
i=0
for key in channels.keys():
    ax = axes[i]
    
    data = channels[key]
    freq = sampling_freq[key]
    
    ax.set_title(key,size=15)
    ax.plot(samplings[freq],data)
    
    i+=1

f.tight_layout()
```
